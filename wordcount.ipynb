```
import findspark
findspark.init()
```

```
# SparkContext
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
```

```
spark = SparkSession.builder\
.master('local[4]')\
.appName('Wordcount_Rdd').getOrCreate()
```

```
sc = spark.sparkContext
```

```
data_file = 'dbfs:/FileStore/shared_uploads/glshyagci@gmail.com/omer_seyfettin_forsa_hikaye.txt'
```

```
story_rdd = sc.textFile(data_file)
```

```
# satırların sayısı
story_rdd.count()
```

```
# 5 satırı alalım.
story_rdd.take(5)
```

```
# transformation
#flatMap = herbir satırı gezer
words = story_rdd.flatMap(lambda row: row.split(' '))
```

```
words.take(5)
```

```
words_count = words.map(lambda each:(each,1))
```

```
words_count.take(5)
```

```
words_count_RBK = words_count.reduceByKey(lambda x,y: x+y)
```

```
words_count_RBK.take(10)
```

```
words_count_RBK2 = words_count_RBK.map(lambda x: (x[1],x[0]))
words_count_RBK2.take(5)
```

```
words_count_RBK2.sortByKey(False).take(20)
```
